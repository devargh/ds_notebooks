# Load the data

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import OrdinalEncoder

# https://www.kaggle.com/datasets/debajyotipodder/co2-emission-by-vehicles
df = pd.read_csv('data/CO2 Emissions_Canada.csv', sep=',')
```

```python
ord_enc = OrdinalEncoder()
ord_enc.fit(df[["Vehicle Class", "Transmission", "Fuel Type"]])
df[["Vehicle Class", "Transmission", "Fuel Type"]] = ord_enc.transform(df[["Vehicle Class", "Transmission", "Fuel Type"]])

df.tail()
```
